```python
import pandas as pd
import numpy as np
```

Reading the Disease-Symptoms dataset into the python notebook

```python
df=pd.read_csv('Dataset2.csv')
```

```python
df.info()
```

Creating the Disease-Specialty mapping

```python
disease_specialist_mapping={}
for i in df.index:
    disease_specialist_mapping[df.loc[i,'Disease']]=df.loc[i,'Specialist']
disease_specialist_mapping
```

Printing the list of unique specialties present in the dataset

```python
print(df['Specialist'].unique())
print(len(df['Specialist'].unique()))
```

Printing the list of diseases present in the dataset

```python
print(df['Disease'].unique())
print(len(df['Disease'].unique()))
```

Printing the list of symptoms present in the dataset

```python
symptoms_list=[]
for i in range(1,18):
    symptoms_list+=(list(df['Symptom_'+str(i)]))

symptoms_list=list(set(symptoms_list))
symptoms_list=symptoms_list[1:]
symptoms_list[3]='itching'
print(len(symptoms_list))
print(symptoms_list)
```

```python
l=symptoms_list.copy()
for i in range(len(l)):
    l[i]=l[i].strip()
l=['Specialist','Disease']+l
dfnew=pd.DataFrame(columns=l)
```

One hot encoding for the symptoms

Preprocessing the dataset for making it suitable for input to the ML model

```python
for i in df.index:
    s=df.loc[i, 'Specialist']
    d=df.loc[i, 'Disease']
    symptoms=[0]*132
    for j in range(1,18):
        print(i,j)
        symp=df.loc[i,'Symptom_'+str(j)]
        if str(symp)!='nan':
            index=symptoms_list.index(symp)
            symptoms[index]=1
    df_temp=pd.DataFrame([[s]+[d]+symptoms],columns=list(dfnew.columns))
    dfnew=dfnew.append(df_temp, ignore_index=True)
dfnew
```

Save the one-hot encoded dataframe into a csv file for future reference

```python
dfnew.to_csv (r'model_train_test.csv', index = False, header=True)
```

```python
dfnew=pd.read_csv('model_train_test.csv')
```

Spliting the dataset into training and testing dataset

```python
dftrain,dftest=pd.DataFrame(),pd.DataFrame()
for d in list(dfnew['Disease'].unique()):
    temp_df=dfnew[dfnew['Disease']==d]
    temp_df.sample(frac=1)
    dftrain=dftrain.append(temp_df[:100])
    dftest=dftest.append(temp_df[100:])
```

```python
specialist_list=list(dftrain['Specialist'].unique())
disease_list=list(dftrain['Disease'].unique())
print(specialist_list)
print(disease_list)
```

```python
x_train,y_train=[],[]
for i in dftrain.index:
    l=list(dftrain.loc[i])
    x_train.append(l[2:])
    y_train.append(disease_list.index(l[1]))
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
print(x_train.shape)
print(y_train.shape)
```

Model Building and training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
model=MLPClassifier((100,50,25,5),learning_rate='adaptive',early_stopping=True,validation_fraction=0.1)
model.fit(x_train,y_train)
```

```python
x_test,y_test=[],[]
for i in dftest.index:
    l=list(dftest.loc[i])
    x_test.append(l[2:])
    y_test.append(disease_list.index(l[1]))

x_test=np.array(x_test)
y_test=np.array(y_test)
print(x_test.shape)
print(y_test.shape)
```

Testing the accuracy of the model

```python
predicted=model.predict(x_test)
```

```python
correct_count=0
for i in range(len(x_test)):
    if predicted[i]==y_test[i]:
        correct_count+=1
print(round(correct_count/len(x_test)*100,3))
```

Comparing models of other classification parameters like precision, recall and f1-score

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted))
```

Saving the model built in a pickle file for future reference

```python
import pickle
pickle.dump(model, open('model.pkl', 'wb'))
```
